```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Guermoud98/iPhone-Sales-Analysis-Project/refs/heads/main/apple_products.csv")
```

# Top 10 Apples Iphone Sales On The Basis of Star Rating

```python
highest_rated = data.sort_values(by = ["Star Rating"], ascending = False)
highest_rated = highest_rated.head(10)
print(highest_rated['Product Name'])
```

# Lets Have a Look at the number of rating of the highest rated iphone on the flipkart

```python
iphones = highest_rated['Product Name'].value_counts()
labels = iphones.index
counts = highest_rated['Number Of Ratings']
figure = px.bar(highest_rated, x=labels, y=counts, title = "Number of Rating of highest related iphone")
figure.show()
```

```python
iphones
```

# Number of Reviews of Highest Rated Iphone

```python
iphones = highest_rated['Product Name']
label = iphones.index
counts = highest_rated['Number Of Reviews']
figure = px.bar(highest_rated, x = labels, y = counts, title = "Number of Reviews of Highest Rated Iphone")
figure.show()
```

# RelationShip between Sale Price and Number of Rating

```python
figure = px.scatter(data_frame = data, x = "Number Of Ratings", 
                    y = "Sale Price", size = "Discount Percentage", trendline = "ols", 
                    title = "RelationShip between sale price and number of ratings")
figure.show()
```

# Relation between discount percentage and number of rating

```python
figure = px.scatter(data_frame = data, x = "Number Of Ratings", y="Discount Percentage", 
                   size = "Sale Price", trendline = "ols",
                   title = "Relation between discount percentage and number of rating")
figure.show()
```
